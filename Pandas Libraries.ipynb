# pandas.melt()

```python
df=pd.DataFrame({'Name':{0: 'John',1: 'Bob',2:'Shiela'},
                 'Course':{0:'Masters',1: 'Graduate',2: 'Graduate'},
                 'Age':{0:27,1:23,2:21}})
df
```

```python
pd.melt(df,id_vars=['Name'],value_vars=['Course'])
```

```python
pd.melt(df,id_vars=['Name'],value_vars=['Course','Age'])
```

# pandas.pivot()

```python
df=pd.DataFrame({'A': ['John','Boby','Mina'],'B':['Masters','Graduate','Graduate'],'C':[27,23,21]})
df
```

```python
df.pivot('A','B','C')
```

```python
df.pivot(index='A',columns='B',values=['C','A'])
```

# pandas.pivot_table()

```python
df=pd.DataFrame({'A':['John','Boby','Mina','Peter','Nicky'],
                'B':['Masters','Graduate','Graduate','Masters','Graduate'],
                'C':[27,23,21,23,24]})
df
```

```python
table=pd.pivot_table(df,index=['A','B'])
table
```

```python
table=pd.pivot_table(df,values='A',index=['B','C'],
                    columns=['B'],aggfunc=np.sum)
table
```

# pandas.merge()

```python
df1=pd.DataFrame({'Name':['Raju','Rani','Geeta','Sita','Sohit'],'Marks':[80,90,75,88,59]})
df2=pd.DataFrame({'Name':['Raju','Divya','Geeta','Sita'],'Grade':['A','A','B','A'], 'Rank':[3,1,4,2],
                 'Gender':['Male','Female','Female','Female']})
display(df1)
display(df2)
df1.merge(df2[['Name','Grade','Rank']])
```

# pandas.merge_ordered()

```python
df1=pd.DataFrame({"date": ['2007-02-01','2007-03-01','2007-04-01','2007-05-01','2007-06-01'],
"close": [12.08, 13.27, 14.27, 17.31, 17.43]})
print("The First DataFrame")
print(df1)

df2=pd.DataFrame({"date":['2007-01-01','2007-02-01','2007-03-01','2007-04-01','2007-05-01'],
"close": [44.34, 43.68, 45.04, 48.27, 50.54]})
print("The Second DataFrame")
print(df2)

print("The merge_ordered DataFrame")
df=pd.merge_ordered(df1,df2, on='date',suffixes=('_df1','_df2'))

print(df)
```

# pandas.concat()

```python
df1=pd.DataFrame(({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}))
display(df1)
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7']})
display(df2)

display('After concatenating:')
display(pd.concat([df1,df2],ignore_index= False ))
```

```python
df3=pd.DataFrame(({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}))
display(df3)
df4=pd.DataFrame({'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7']})
display(df4)

display('After concatenating:')
display(pd.concat([df3,df4],axis=1 ))
```

```python
df5=pd.DataFrame(({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}))
display(df5)
df6=pd.DataFrame({'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7']})
display(df6)

display('After concatenating:')
display(pd.concat([df5,df6],keys=['key1','key2']))
```

```python
df7=pd.DataFrame(({'Roll no':[1,2,3,4],'Name':['Hari','Abi','Kishore','Vishnu']}))
display(df7)
df8=pd.DataFrame({'Roll no':[1,2,7,8],'Name':['Lavanya','Sivaranjani','Jaishree','Madhu']})
display(df8)

display('After concatenating:')
display(pd.concat([df7,df8],ignore_index= True ))
```

# pandas.get_dummies() - one hot encoding

```python
df=pd.DataFrame({'Name':['John','Chandler','Joey','Ross','Rachel'],
                'Gender':['Male','Male','Male','Male','Female'],
                'House Type':['Apartment','Detached','Apartment',None,'Semi-Detached']})
display(df)
dummie=pd.get_dummies(data=df,columns=['Gender'])
display(dummie)
```

# pandas.from_dummies()

```python

d=pd.DataFrame({'a':[1,0,0,1],'b':[0,1,0,0],'c':[0,0,1,0]})
print(d)
pd.Categorical(d)
```

```python
# Create a DataFrame with dummy-coded variables
df = pd.DataFrame({'a':[1,0,0,1],'b':[0,1,0,0],'c':[0,0,1,0]})
print("Dummy-coded DataFrame:")
print(df)

# Convert dummy-coded variables back to a categorical variable
result = df.idxmax(axis=1)
print("\nCategorical variable:")
print(result)
```

# pandas.factorize()

```python
df=pd.DataFrame({'conf':['West','West','East','East'],
                'team':['A','B','C','D'],
                'position':['Guard','Forward','Guard','Center']})
df
```

```python
df['conf']=pd.factorize(df['conf'])[0]
df
```

```python
df[['conf','team']]=df[['conf','team']].apply(lambda x: pd.factorize(x)[0])
df
```

```python
df=df.apply(lambda x: pd.factorize(x)[0])
df
```

# pandas.wide_to_long()

```python
df=pd.DataFrame({"car_names":['Suzuki','Honda','Toyota'],
                "fuel_type":['diesel','petrol','diesel'],
                "jan23":[342,432,123],
                "jan22":[245,543,334],
                "jan21":[544,209,324]})
print(df)
pd.wide_to_long(df,stubnames='jan',i='car_names',j='year')
```

```python
pd.wide_to_long(df,stubnames='jan',i=['car_names','fuel_type'],j='year')
```

```python
jandf=pd.DataFrame({"city":['Mumbai','Delhi','Pune'],
                   "January-1":[21,16,19],
                   "January-2":[25,20,22],
                   "January-3":[20,18,19]})
display(jandf)
pd.wide_to_long(jandf,stubnames='January',i='city',j='day',sep='-')
```

# pandas.eval() used to calculate row wise

```python
print(pd.eval("2+3"))
print(pd.eval("2+3*(5-2)"))
```

```python
data=pd.DataFrame({"Student":["A","B","C","D"],"Physics":[89,34,23,56],"Chemistry":[34,56,98,56],"Maths":[34,94,50,59]})
display(data)
data['Total']=pd.eval("data.Physics+data.Chemistry+data.Maths")
display(data)
pd.eval("Avg=data.Total/3",target=data,inplace=True)
display(data)
```
